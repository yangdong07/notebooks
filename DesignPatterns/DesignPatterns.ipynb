## 基本概念


### 设计模式六大原则

#### 开闭原则（Open Close Principle）

开闭原则就是说对扩展开放，对修改关闭。在程序需要进行拓展的时候，不能去修改原有的代码，实现一个热插拔的效果。

注解：开闭原则需要“很完美”的抽象概括能力，支持扩展开发，但是禁止修改。所以如果抽象接口设计的不好，其实还是很不方便的。

#### 里氏代换原则（Liskov Substitution Principle）

里氏代换原则 (Liskov Substitution Principle LSP) 面向对象设计的基本原则之一。 里氏代换原则中说，任何基类可以出现的地方，子类一定可以出现。 LSP是继承复用的基石，只有当衍生类可以替换掉基类，软件单位的功能不受到影响时，基类才能真正被复用，而衍生类也能够在基类的基础上增加新的行为。里氏代换原则是对“开-闭”原则的补充。实现“开-闭”原则的关键步骤就是抽象化。而基类与子类的继承关系就是抽象化的具体实现，所以里氏代换原则是对实现抽象化的具体步骤的规范。

#### 依赖倒转原则（Dependence Inversion Principle）
这个是开闭原则的基础，具体内容：针对接口编程，依赖于抽象而不依赖于具体。

注解：其实通俗点说，就是守本分。做好自己的事情。

#### 接口隔离原则（Interface Segregation Principle）

这个原则的意思是：使用多个隔离的接口，比使用单个接口要好。还是一个降低类之间的耦合度的意思，从这儿我们看出，其实设计模式就是一个软件的设计思想，从大型软件架构出发，为了升级和维护方便。所以上文中多次出现：降低依赖，降低耦合。

注解：这个原则背后的思想是，将任务颗粒度变得更低。即将问题分解的更细致。 

#### 迪米特法则（最少知道原则）（Demeter Principle）
为什么叫最少知道原则，就是说：一个实体应当尽量少的与其他实体之间发生相互作用，使得系统功能模块相对独立。

注解：意图是减少交互关系、减少依赖关系。

#### 合成复用原则（Composite Reuse Principle）

原则是尽量使用合成/聚合的方式，而不是使用继承。

注解： 其实继承是很僵硬的关系。不够灵活。


### 类之间的关系

1. 泛化关系(generalization)。 泛化关系表现为继承非抽象类。泛化关系用一条带空心箭头的直接表示
2. 实现关系(realize)。实现关系表现为继承抽象类。实现关系用一条带空心箭头的虚线表示；
3. 聚合关系(aggregation)。聚合关系用于表示实体对象之间的关系，表示整体由部分构成的语义；例如一个部门由多个员工组成。整体和部分不是强依赖的。聚合关系用一条带空心菱形箭头的直线表示。
4. 组合关系(composition)。组合关系是一种强依赖的特殊聚合关系。用一条带实心菱形箭头的直线表示
5. 关联关系(association)。关联关系是用一条直线表示的；它描述不同类的对象之间的结构关系；它是一种静态关系， 通常与运行状态无关，一般由常识等因素决定的；它一般用来定义对象之间静态的、天然的结构； 所以，关联关系是一种“强关联”的关系；比如，乘车人和车票之间就是一种关联关系；学生和学校就是一种关联关系；在最终代码中，关联对象通常是以成员变量的形式实现的；
6. 依赖关系(dependency)。依赖关系是用一套带箭头的虚线表示的；


6个关系， 泛化和实现是继承的两种关系，泛化是继承非抽象类，例如 SUV类继承汽车类。 用空心箭头直线表示。 实现是继承抽象类，例如 汽车类继承车接口类。用空心箭头虚线表示。

聚合关系和组合关系。 聚合是弱依赖关系：乌合之众， 用空心菱形箭头表示； 组合是强依赖关系：唇亡齿寒，用实心菱形箭头表示

依赖关系，用带箭头的虚线表示。最好使用单向依赖，避免使用双向依赖。

除去以上明确关系，如果对象之间有一些关联，例如 乘车人和车票、学生和学校，则使用 关联关系，用带箭头直线表示。



### **自己理解的**

模式只是一个基础，如果要灵活应用模式，还需要多学习和多尝试，认识到问题和事物的本质，多积累解决问题的方法。


**所有设计模式，所说不过是"去耦合"三个字。** 在学习设计模式的时候，主要看其思想，以及解决的问题的本质。

模式并不是必要的。 有很多很棒的模式（比如装饰器模式、迭代器模式），在python中很常见，使用起来也很自然。 软件工程的本质是分解、去耦合。一切模式不过是这种思想的具体体现罢了。

## 创建型模式 Creational patterns

> deal with object creation mechanisms, trying to create objects in a manner suitable to the situation. The basic form of object creation could result in design problems or added complexity to the design. Creational design patterns solve this problem by somehow controlling this object creation.

创造对象用的模式： 希望不了解对象产生的机制，仅仅创建对象，然后直接调用。

解决的主要问题，就是 分离创建对象和使用对象。 使用者不用关心对象的创建过程。

### 简单工厂模式 ☆☆☆☆

核心思想：将类的创建过程与对象的使用过程分离。

具体实现过程：定义一个工厂，根据参数（通常是字符串）的不同返回不同类的实例。因为我们希望将工厂生产的对象可以使用相同的方式处理，所以产品类类通常会实现相同的接口。

下面举例说明使用参数创建不同实例。


```python
class GreekGetter(object):

    """A simple localizer a la gettext"""

    def __init__(self):
        self.trans = dict(dog="σκύλος", cat="γάτα")

    def get(self, msgid):
        """We'll punt if we don't have a translation"""
        return self.trans.get(msgid, str(msgid))


class EnglishGetter(object):

    """Simply echoes the msg ids"""

    def get(self, msgid):
        return str(msgid)


def get_localizer(language="English"):
    """The factory method"""
    languages = dict(English=EnglishGetter, Greek=GreekGetter)
    return languages[language]()


if __name__ == '__main__':
    # Create our localizers
    e, g = get_localizer(language="English"), get_localizer(language="Greek")
    # Localize some text
    for msgid in "dog parrot cat bear".split():
        print(e.get(msgid), g.get(msgid))
```


### 工厂方法模式（Factory Method） ☆☆☆☆☆

核心思想：简单工厂基础上，通过实例化不同的工厂，来创建不同的类。

具体实现过程：在简单工厂的基础上，对工厂进行抽象，将类的创建过程延迟到工厂的子类中完成。这样做的好处是当需要新增加一个新的产品类的时候，只需要新增一个“具体的工厂”就可以，而不需要像简单工厂一样，修改工厂类的具体实现。符合“开闭原则”。

注解：从使用者角度：不想知道对象创建的具体过程，只需要知道由哪个工厂创建哪个对象就好了； 从创建者角度：由不同工厂创建不同对象，每种工厂全权负责自己创造的那种对象。

```python
import abc


# 抽象工厂，以及工厂方法接口
class Creator(metaclass=abc.ABCMeta):
    def __init__(self):
        self.product = self._factory_method()

    @abc.abstractmethod
    def _factory_method(self):
        pass

    def some_operation(self):
        self.product.interface()

# 具体工厂1
class ConcreteCreator1(Creator):
    def _factory_method(self):
        return ConcreteProduct1()

# 具体工厂2
class ConcreteCreator2(Creator):
    def _factory_method(self):
        return ConcreteProduct2()

# 抽象产品类，及接口
class Product(metaclass=abc.ABCMeta): 
    @abc.abstractmethod
    def interface(self):
        pass

# 具体产品1
class ConcreteProduct1(Product): 
    def interface(self):
        pass

# 具体产品2
class ConcreteProduct2(Product): 
    def interface(self):
        pass


def main():
    concrete_creator = ConcreteCreator1()
    concrete_creator.product.interface()
    concrete_creator.some_operation()


if __name__ == "__main__":
    main()
```


### 抽象工厂模式（Abstract Factory） ☆☆☆☆☆


核心思想：工厂方法基础上，让一个工厂生产一类对象。

具体实现过程：在工厂方法的基础上，在一个具体工厂中实现生产不同产品的方法。可以达到方便生产一个“产品簇”的目的。


注解： 

创建者角度： 抽象工厂的接口是创建一组产品（产品簇）； 具体工厂实现创建一组产品。（就比如家电工厂创建冰箱、洗衣机； 海尔家电工厂创建海尔牌子的冰箱、洗衣机）

使用者角度： 选择使用某种产品，同时还要选择使用哪个工厂创建。（就比如我要使用一个冰箱，我还要选择是海尔工厂创建的冰箱）

```python
import abc

# 抽象工厂，
class AbstractFactory(metaclass=abc.ABCMeta):
    @abc.abstractmethod
    def create_product_a(self):
        pass

    @abc.abstractmethod
    def create_product_b(self):
        pass

# 具体工厂
class ConcreteFactory1(AbstractFactory):
    def create_product_a(self):
        return ConcreteProductA1()

    def create_product_b(self):
        return ConcreteProductB1()


# 具体工厂
class ConcreteFactory2(AbstractFactory):
    def create_product_a(self):
        return ConcreteProductA2()

    def create_product_b(self):
        return ConcreteProductB2()

# 抽象产品A，代表同类型的产品A
class AbstractProductA(metaclass=abc.ABCMeta):
    @abc.abstractmethod
    def interface_a(self):
        pass

# 具体产品A1
class ConcreteProductA1(AbstractProductA):
    def interface_a(self):
        pass

# 具体产品A2
class ConcreteProductA2(AbstractProductA):
    def interface_a(self):
        pass

# 抽象产品B
class AbstractProductB(metaclass=abc.ABCMeta):
    @abc.abstractmethod
    def interface_b(self):
        pass

# 具体产品B1
class ConcreteProductB1(AbstractProductB):
    def interface_b(self):
        pass

# 具体产品B2
class ConcreteProductB2(AbstractProductB):
    def interface_b(self):
        pass


def main():
    for factory in (ConcreteFactory1(), ConcreteFactory2()):
        product_a = factory.create_product_a()
        product_b = factory.create_product_b()
        product_a.interface_a()
        product_b.interface_b()


if __name__ == "__main__":
    main()
```

### 建造者模式（Builder）  ☆☆

核心思想：定义一个建造过程，具体的建造放在子类当中，这样可以使用相同的建造过程构造不同的实例。

注解： 其实本质上是一个模板方法模式。 这是工厂的建造方法的更具体的实现。

下面这个例子中，分解了建造一个完整对象的步骤。具体的实现放到子类中。 在Director类中组合这些步骤。

```python
import abc


class Director:
    def __init__(self):
        self._builder = None

    def construct(self, builder):
        self._builder = builder
        self._builder._build_part_a()
        self._builder._build_part_b()
        self._builder._build_part_c()


class Builder(metaclass=abc.ABCMeta):
    def __init__(self):
        self.product = Product()

    @abc.abstractmethod
    def _build_part_a(self):
        pass

    @abc.abstractmethod
    def _build_part_b(self):
        pass

    @abc.abstractmethod
    def _build_part_c(self):
        pass


class ConcreteBuilder(Builder):
    def _build_part_a(self):
        pass

    def _build_part_b(self):
        pass

    def _build_part_c(self):
        pass


class Product:
    pass


def main():
    concrete_builder = ConcreteBuilder()
    director = Director()
    director.construct(concrete_builder)
    product = concrete_builder.product


if __name__ == "__main__":
    main()
```

### 单例模式（Singleton） ☆☆☆☆

核心思想：让一个类自身持有其实例的成员变量，声明为`static`和`private`，来达到系统中只有一个实例的目的。

具体实现过程：为了确保全系统只有一个实例，所以不可以让用户自己创建，所以这个类需要一个`private`的构造函数。还需要一个`getInstance()`方法用来获得这个唯一的实例，因为需要在系统的每一处都能获得，所以这个方法应该是`public`的。因为调用的`getInstance()`的目的是获取实例，只能在类上面调用，所以`getInstance()`应该是static的。


下面这个例子用到了 python 的 metaclass。 注意几点即可：

1. metaclass Singletion 是元类，继承自 type， 其实例是类。 所以元类的 `__init__`，其实就是实例化一个元类，即创建一个类。
2. 由类维护一个唯一实例 `_instance`
3. 元类的 `__call__` 可以理解成是 创建类实例的过程（包括类的 `__new__` 和 `__init__`）。如果这唯一的实例存在，则返回，否则就创建一个唯一的。

```python
class Singleton(type):
    def __init__(cls, name, bases, attrs, **kwargs):
        super().__init__(name, bases, attrs)
        cls._instance = None

    def __call__(cls, *args, **kwargs):
        if cls._instance is None:
            cls._instance = super().__call__(*args, **kwargs)
        return cls._instance


class MyClass(metaclass=Singleton):
    pass


def main():
    m1 = MyClass()
    m2 = MyClass()
    assert m1 is m2


if __name__ == "__main__":
    main()
```

### 适配器模式（Adaper）☆☆☆

核心思想：创建一个适配器类可以同时具备有现有类的行为，和期望的接口。

具体实现过程：分为两种情况

1. 类适配器：适配器类实现期望的接口，继承现有的类已获得他的行为（方法）。
2. 对象适配器：适配器类实现期望的接口，并持有现有类的实例，通过实例调用方法。

目的都是相同的。

注解： 适配器可以联想到电源适配器。从用户角度看，期望使用一种接口（比如国标插座），但是只提供了其他接口（比如英标插座），这时候需要一个适配器去调整一下。

这里假定了用户不关心原来的接口是怎么实现的，也不关心现有类的其他作用。 应用面比较窄。

下面的例子是一个对象适配器。 持有现有类的实例，通过实例调用方法

```python
import abc


# 用户期望的接口
class Target(metaclass=abc.ABCMeta):
    def __init__(self):
        self._adaptee = Adaptee()

    @abc.abstractmethod
    def request(self):
        pass

# 适配器
class Adapter(Target):
    def request(self):
        self._adaptee.specific_request()

# 现有类，特殊接口
class Adaptee:
    def specific_request(self):
        pass


def main():
    adapter = Adapter()
    adapter.request()


if __name__ == "__main__":
    main()
```

### 桥接模式（Bridge） ☆☆☆

核心思想：使用组合或者聚合的方式处理抽象和实现之间的联系，让具体的实现在不同维度上变化

注解： 将抽象和实现的继承关系改成了 关联关系。

这样考虑桥接模式

- 从实现者（Implementor）角度看，需要实现一系列基本操作（primitive operations），不同的实现者实现方式不同（例如视频解码）。这是一个维度上的变化；
- 从实现者的抽象类（Abstraction）角度看，想要对实现者的接口方法进行抽象，但是实现者还没有定型。我仍然可以定义一些接口，调用实现者的实例进行操作。 这是另一个维度上的变化。
- 从使用者角度看， 并不关心实现者背后复杂的实现，只关心抽象接口。 **使用者只关心接口**

很像适配器，但是允许适配者灵活变化，同时自己的接口也可以发生改变。 这就是通过组合或者聚合的方式处理抽象和实现之间的联系。 将抽象和实现的继承关系改成了 关联关系。

```python
import abc


class Abstraction:
    """
    Define the abstraction's interface.
    Maintain a reference to an object of type Implementor.
    """

    def __init__(self, imp):
        self._imp = imp

    def operation(self):
        self._imp.operation_imp()


        
class Implementor(metaclass=abc.ABCMeta):
    """
    Define the interface for implementation classes. This interface
    doesn't have to correspond exactly to Abstraction's interface; in
    fact the two interfaces can be quite different. Typically the
    Implementor interface provides only primitive operations, and
    Abstraction defines higher-level operations based on these
    primitives.
    """

    @abc.abstractmethod
    def operation_imp(self):
        pass


class ConcreteImplementorA(Implementor):
    """
    Implement the Implementor interface and define its concrete
    implementation.
    """

    def operation_imp(self):
        pass


class ConcreteImplementorB(Implementor):
    """
    Implement the Implementor interface and define its concrete
    implementation.
    """

    def operation_imp(self):
        pass


def main():
    concrete_implementor_a = ConcreteImplementorA()
    abstraction = Abstraction(concrete_implementor_a)
    abstraction.operation()


if __name__ == "__main__":
    main()
```

### 组合模式（Composite）☆☆☆

核心思想：让具体类和这些类的容器实现同一个接口，用户调用到容器类时使用递归调用，不需要区分这两者的区别。


注解： 可以稍微这样理解： 个人的是集体的，集体的还是集体的。 

```python
import abc


# 组件基类（抽象类）
class Component(metaclass=abc.ABCMeta):
    """
    Declare the interface for objects in the composition.
    Implement default behavior for the interface common to all classes,
    as appropriate.
    Declare an interface for accessing and managing its child
    components.
    Define an interface for accessing a component's parent in the
    recursive structure, and implement it if that's appropriate
    (optional).
    """

    @abc.abstractmethod
    def operation(self):
        pass

# 容器类（由组件组成）
class Composite(Component):
    """
    Define behavior for components having children.
    Store child components.
    Implement child-related operations in the Component interface.
    """

    def __init__(self):
        self._children = set()

    def operation(self):
        for child in self._children:
            child.operation()

    def add(self, component):
        self._children.add(component)

    def remove(self, component):
        self._children.discard(component)

# 组件具体类
class Leaf(Component):
    """
    Represent leaf objects in the composition. A leaf has no children.
    Define behavior for primitive objects in the composition.
    """

    def operation(self):
        pass


def main():
    leaf = Leaf()
    composite = Composite()
    composite.add(leaf)
    composite.operation()


if __name__ == "__main__":
    main()
```

### 装饰模式（Decorator）☆☆☆☆☆

核心思想： **装饰器和被装饰的对象有相同的接口，装饰器在使用时获得被装饰的对象，为其增加新的操作。**

注解： 在python中很常见。 decorator。  注意一点： 接口相同，但此人非彼人也。

下面这个例子与python的装饰器不同，教科书做法。python自带的装饰器语法更精简。领会其中精神即可

```python
import abc


# 组件抽象类
class Component(metaclass=abc.ABCMeta):
    @abc.abstractmethod
    def operation(self):
        pass

# 装饰器类
class Decorator(Component, metaclass=abc.ABCMeta):
    def __init__(self, component):
        self._component = component

    @abc.abstractmethod
    def operation(self):
        pass

# 具体装饰器A
class ConcreteDecoratorA(Decorator):
    """
    Add responsibilities to the component.
    """

    def operation(self):
        # ...
        self._component.operation()
        # ...


# 具体装饰器B
class ConcreteDecoratorB(Decorator):
    """
    Add responsibilities to the component.
    """

    def operation(self):
        # ...
        self._component.operation()
        # ...

# 具体组件
class ConcreteComponent(Component):
    def operation(self):
        pass


def main():
    concrete_component = ConcreteComponent()
    concrete_decorator_a = ConcreteDecoratorA(concrete_component)
    concrete_decorator_b = ConcreteDecoratorB(concrete_decorator_a)
    concrete_decorator_b.operation()


if __name__ == "__main__":
    main()
```

### 外观模式（Facade）☆☆☆☆

核心思想：**在外观类内实现具体的操作，用户只需要和外观类打交道，屏蔽方法调用细节。**


注解： 很好理解。从使用者角度，还是不关心细节的调用。细节交给 `Facade` 对象处理。 缺点是，不够灵活，使用者如果关心内部细节实现并且希望干预，就不够灵活。

```python
class Facade:
    def __init__(self):
        self._subsystem_1 = Subsystem1()
        self._subsystem_2 = Subsystem2()

    def operation(self):
        self._subsystem_1.operation1()
        self._subsystem_1.operation2()
        self._subsystem_2.operation1()
        self._subsystem_2.operation2()


class Subsystem1:
    def operation1(self):
        pass

    def operation2(self):
        pass


class Subsystem2:
    def operation1(self):
        pass

    def operation2(self):
        pass


def main():
    facade = Facade()
    facade.operation()


if __name__ == "__main__":
    main()
```

### 享元模式（Flyweight）☆

核心思想：建一个对象池，复用需要在系统中多次出现的对象。

注解： 享元模式是一个考虑系统性能的设计模式，通过使用享元模式可以节约内存空间，提高系统的性能。

享元模式以共享的方式高效地支持大量的细粒度对象，享元对象能做到共享的关键是区分内部状态(Internal State)和外部状态(External State)。

但是缺点是： 享元模式使得系统更加复杂，需要分离出内部状态和外部状态，这使得程序的逻辑复杂化。

```python
import abc


class FlyweightFactory:
    """
    Create and manage flyweight objects.
    Ensure that flyweights are shared properly. When a client requests a
    flyweight, the FlyweightFactory object supplies an existing instance
    or creates one, if none exists.
    """

    def __init__(self):
        self._flyweights = {}

    def get_flyweight(self, key):
        try:
            flyweight = self._flyweights[key]
        except KeyError:
            flyweight = ConcreteFlyweight()
            self._flyweights[key] = flyweight
        return flyweight


class Flyweight(metaclass=abc.ABCMeta):
    """
    Declare an interface through which flyweights can receive and act on
    extrinsic state.
    """

    def __init__(self):
        self.intrinsic_state = None

    @abc.abstractmethod
    def operation(self, extrinsic_state):
        pass


class ConcreteFlyweight(Flyweight):
    """
    Implement the Flyweight interface and add storage for intrinsic
    state, if any. A ConcreteFlyweight object must be sharable. Any
    state it stores must be intrinsic; that is, it must be independent
    of the ConcreteFlyweight object's context.
    """

    def operation(self, extrinsic_state):
        pass


def main():
    flyweight_factory = FlyweightFactory()
    concrete_flyweight = flyweight_factory.get_flyweight("key")
    concrete_flyweight.operation(None)


if __name__ == "__main__":
    main()
```

### 代理模式（Proxy）☆☆☆☆☆

核心思想：代理类持有具体实现类的实例，将在代理类上的操作转化为实例上方法的调用。

注解：代理模式与装饰器模式非常像。 代理模式的一个关键是： 接口相同，间接调用。 代理的最大特点在于访问控制。 装饰器模式的最大特点是添加功能。 装饰器普通、灵活、常见。 代理显得比较专业。 装饰器是狸猫换太子， 代理是职业中介

```python
import abc


class Subject(metaclass=abc.ABCMeta):
    """
    Define the common interface for RealSubject and Proxy so that a
    Proxy can be used anywhere a RealSubject is expected.
    """

    @abc.abstractmethod
    def request(self):
        pass


class Proxy(Subject):
    """
    Maintain a reference that lets the proxy access the real subject.
    Provide an interface identical to Subject's.
    """

    def __init__(self, real_subject):
        self._real_subject = real_subject

    def request(self):
        # ...
        self._real_subject.request()
        # ...


class RealSubject(Subject):
    """
    Define the real object that the proxy represents.
    """

    def request(self):
        pass


def main():
    real_subject = RealSubject()
    proxy = Proxy(real_subject)
    proxy.request()


if __name__ == "__main__":
    main()
```

## 行为型模式（Behavioral patterns）

一句话总结设计模式的[文章](https://www.jianshu.com/p/dd2bbbbcddab)，但是没有对行为型模式的说明，甚是遗憾！


行为型模式(Behavioral Pattern)是对在不同的对象之间划分责任和算法的抽象化。

行为型模式不仅仅关注类和对象的结构，而且重点关注它们之间的相互作用。

通过行为型模式，可以更加清晰地划分类与对象的职责，并研究系统在运行时实例对象 之间的交互。在系统运行时，对象并不是孤立的，它们可以通过相互通信与协作完成某些复杂功能，一个对象在运行时也将影响到其他对象的运行。



### 命令模式（command） ☆☆☆


核心思想： 命令模式的本质是对命令进行封装，将发出命令的责任和执行命令的责任分割开。


命令模式(Command Pattern)：将一个请求封装为一个对象，从而使我们可用不同的请求对客户进行参数化；对请求排队或者记录请求日志，以及支持可撤销的操作。命令模式是一种对象行为型模式，其别名为动作(Action)模式或事务(Transaction)模式。


注解： 命令模式的一个使用场景是，系统需要将请求调用者和请求接收者解耦，使得调用者和接收者不直接交互。 本质上是 分离了动作的调用者和动作的执行者。


```python
import abc


class Invoker:
    """
    Ask the command to carry out the request.
    """

    def __init__(self):
        self._commands = []

    def store_command(self, command):
        self._commands.append(command)

    def execute_commands(self):
        for command in self._commands:
            command.execute()

        
# 命令基类， 定义执行接口
class Command(metaclass=abc.ABCMeta):
    """
    Declare an interface for executing an operation.
    """

    def __init__(self, receiver):
        self._receiver = receiver

    @abc.abstractmethod
    def execute(self):
        pass

    
# 具体命令
class ConcreteCommand(Command):
    """
    Define a binding between a Receiver object and an action.
    Implement Execute by invoking the corresponding operation(s) on
    Receiver.
    """

    def execute(self):
        self._receiver.action()


class Receiver:
    """
    Know how to perform the operations associated with carrying out a
    request. Any class may serve as a Receiver.
    """

    def action(self):
        pass


def main():
    receiver = Receiver()
    concrete_command = ConcreteCommand(receiver)
    invoker = Invoker()
    invoker.store_command(concrete_command)
    invoker.execute_commands()


if __name__ == "__main__":
    main()
```


### 中介者模式（Mediator）☆☆☆☆

中介者模式(Mediator Pattern)定义：用一个中介对象来封装一系列的对象交互，中介者使各对象不需要显式地相互引用，从而使其耦合松散，而且可以独立地改变它们之间的交互。中介者模式又称为调停者模式，它是一种对象行为型模式。

注解：中介者模式是一种行为模式，本质上，是避免多个对象之间的直接交互，过度耦合。又称调停者模式，可以理解为 调度者。 比如火车调度，火车之间不直接通信，而是统一调度。 如果事务过多，中介者模式会显得复杂，但是是有必要的。

```python
# 中介者
class Mediator:
    def __init__(self):
        self._colleague_1 = Colleague1(self)
        self._colleague_2 = Colleague2(self)

# 同事1，只认识中介
class Colleague1:
    def __init__(self, mediator):
        self._mediator = mediator

# 同事2，只认识中介
class Colleague2:
    def __init__(self, mediator):
        self._mediator = mediator


def main():
    mediator = Mediator()


if __name__ == "__main__":
    main()
```

###  观察者模式（Observer） ☆☆☆☆☆

观察者模式(Observer Pattern)：定义对象间的一种一对多依赖关系，使得每当一个对象状态发生改变时，其相关依赖对象皆得到通知并被自动更新。观察者模式又叫做发布-订阅（Publish/Subscribe）模式、模型-视图（Model/View）模式、源-监听器（Source/Listener）模式或从属者（Dependents）模式。

注解： 非常常见的模式之一。 又称 发布-订阅模式，更好理解。 需要注意一点： 订阅者的变化是从被观察对象的主体发起的。如果是同步操作，并且订阅者很多的话，会影响性能。 

```python
import abc

# 被观察者对象
class Subject:
    """
    Know its observers. Any number of Observer objects may observe a
    subject.
    Send a notification to its observers when its state changes.
    """

    def __init__(self):
        self._observers = set()
        self._subject_state = None

    def attach(self, observer):
        observer._subject = self
        self._observers.add(observer)

    def detach(self, observer):
        observer._subject = None
        self._observers.discard(observer)

    def _notify(self):
        for observer in self._observers:
            observer.update(self._subject_state)

    @property
    def subject_state(self):
        return self._subject_state

    @subject_state.setter
    def subject_state(self, arg):
        self._subject_state = arg
        self._notify()


# 观察者抽象类，主要提供一个update方法接口
class Observer(metaclass=abc.ABCMeta):
    """
    Define an updating interface for objects that should be notified of
    changes in a subject.
    """

    def __init__(self):
        self._subject = None
        self._observer_state = None

    @abc.abstractmethod
    def update(self, arg):
        pass


# 具体观察者
class ConcreteObserver(Observer):
    """
    Implement the Observer updating interface to keep its state
    consistent with the subject's.
    Store state that should stay consistent with the subject's.
    """

    def update(self, arg):
        self._observer_state = arg
        # ...


def main():
    subject = Subject()
    concrete_observer = ConcreteObserver()
    subject.attach(concrete_observer)
    subject.subject_state = 123


if __name__ == "__main__":
    main()
```


### 状态模式（State）☆☆☆

状态模式(State Pattern) ：允许一个对象在其内部状态改变时改变它的行为，对象看起来似乎修改了它的类。其别名为状态对象(Objects for States)，状态模式是一种对象行为型模式。

注解： 由于历史原因，以前对状态模式研究比较多。。 对于简单状态机，简单描述状态迁移即可； 对于复杂状态机，可以将状态定义为对象。状态模式描述了对象状态的变化以及对象如何在每一种状态下表现出不同的行为。

实际中的状态模式比下面例子更加复杂。

```python
import abc


# 保存和维护状态信息
class Context:
    def __init__(self, state):
        self._state = state

    def request(self):
        self._state.handle()

# 状态基类（一般情况下会有很多动作）
class State(metaclass=abc.ABCMeta):
    """
    Define an interface for encapsulating the behavior associated with a
    particular state of the Context.
    """

    @abc.abstractmethod
    def handle(self):
        pass

# 具体状态A
class ConcreteStateA(State):
    def handle(self):
        pass

# 具体状态B
class ConcreteStateB(State):
    def handle(self):
        pass


def main():
    concrete_state_a = ConcreteStateA()
    context = Context(concrete_state_a)
    context.request()


if __name__ == "__main__":
    main()
```


### 策略模式（Strategy）☆☆☆

策略模式(Strategy Pattern)：定义一系列算法，将每一个算法封装起来，并让它们可以相互替换。策略模式让算法独立于使用它的客户而变化，也称为政策模式(Policy)。

注解： 这个也很好理解。 从策略实现者角度： 准备好一系列算法； 从使用者角度，选择合适的算法执行就行了。

```python
import abc


# 使用策略的接口
class Context:
    """
    Define the interface of interest to clients.
    Maintain a reference to a Strategy object.
    """

    def __init__(self, strategy):
        self._strategy = strategy

    def context_interface(self):
        self._strategy.algorithm_interface()


# 策略基类
class Strategy(metaclass=abc.ABCMeta):
    """
    Declare an interface common to all supported algorithms. Context
    uses this interface to call the algorithm defined by a
    ConcreteStrategy.
    """

    @abc.abstractmethod
    def algorithm_interface(self):
        pass

    
# 具体策略A
class ConcreteStrategyA(Strategy):
    """
    Implement the algorithm using the Strategy interface.
    """

    def algorithm_interface(self):
        pass

    
# 具体策略B
class ConcreteStrategyB(Strategy):
    """
    Implement the algorithm using the Strategy interface.
    """

    def algorithm_interface(self):
        pass

def main():
    concrete_strategy_a = ConcreteStrategyA()
    context = Context(concrete_strategy_a)
    context.context_interface()


if __name__ == "__main__":
    main()
```


### 模板方法模式（Template Method）☆☆


注解： 也是很简单的模式，在基类里面定义好模板， 子类继承实现模板中需要的步骤方法就好了。 本质上，是分离了行为框架和具体的行为。  跟建造者模式有点像，但是建造者的目的是构建一个完整对象。


个人感觉，用起来还是比较不方便的： 一旦限定了基类模板，子类就必须按照这个模板来。 不是很灵活。 但如果行为有规范说明（必须符合什么样的流程步骤），这里就比较适用。但是这种情况出现的比较少。

```python
import abc


class AbstractClass(metaclass=abc.ABCMeta):
    """
    Define abstract primitive operations that concrete subclasses define
    to implement steps of an algorithm.
    Implement a template method defining the skeleton of an algorithm.
    The template method calls primitive operations as well as operations
    defined in AbstractClass or those of other objects.
    """

    def template_method(self):
        self._primitive_operation_1()
        self._primitive_operation_2()

    @abc.abstractmethod
    def _primitive_operation_1(self):
        pass

    @abc.abstractmethod
    def _primitive_operation_2(self):
        pass


class ConcreteClass(AbstractClass):
    """
    Implement the primitive operations to carry out
    subclass-specificsteps of the algorithm.
    """

    def _primitive_operation_1(self):
        pass

    def _primitive_operation_2(self):
        pass


def main():
    concrete_class = ConcreteClass()
    concrete_class.template_method()


if __name__ == "__main__":
    main()
```

### 访问者模式（Visitor）☆☆☆☆

在访问者模式（Visitor Pattern）中，我们使用了一个访问者类，它改变了元素类的执行算法。通过这种方式，元素的执行算法可以随着访问者改变而改变。这种类型的设计模式属于行为型模式。根据模式，元素对象已接受访问者对象，这样访问者对象就可以处理元素对象上的操作。


注解： 访问者模式的意图是将数据结构与数据操作分离： 操作数据的就是访问者。 不同的访问者会有不同操作方法； 数据结构就是元素类，有不同实现。 感觉比较奇怪，正常思路是：不同数据不同操作方法，由操作方法决定去操作哪种数据； 但是访问者模式是反过来： 由数据决定使用哪种操作方法（访问）。其实二者都有。

不同数据不同操作方法，这是经常遇到的一个场景。 所以访问者模式也比较常见。

下面这个例子说明： 由数据决定使用哪个访问者。

```python
import abc


class Element(metaclass=abc.ABCMeta):
    @abc.abstractmethod
    def accept(self, visitor):
        pass


class ConcreteElementA(Element):
    """
    Implement an Accept operation that takes a visitor as an argument.
    """

    def accept(self, visitor):
        visitor.visit_concrete_element_a(self)


class ConcreteElementB(Element):
    """
    Implement an Accept operation that takes a visitor as an argument.
    """

    def accept(self, visitor):
        visitor.visit_concrete_element_b(self)


class Visitor(metaclass=abc.ABCMeta):
    """
    Declare a Visit operation for each class of ConcreteElement in the
    object structure. The operation's name and signature identifies the
    class that sends the Visit request to the visitor. That lets the
    visitor determine the concrete class of the element being visited.
    Then the visitor can access the element directly through its
    particular interface.
    """

    @abc.abstractmethod
    def visit_concrete_element_a(self, concrete_element_a):
        pass

    @abc.abstractmethod
    def visit_concrete_element_b(self, concrete_element_b):
        pass


class ConcreteVisitor1(Visitor):
    """
    Implement each operation declared by Visitor. Each operation
    implements a fragment of the algorithm defined for the corresponding
    class of object in the structure. ConcreteVisitor provides the
    context for the algorithm and stores its local state. This state
    often accumulates results during the traversal of the structure.
    """

    def visit_concrete_element_a(self, concrete_element_a):
        pass

    def visit_concrete_element_b(self, concrete_element_b):
        pass


class ConcreteVisitor2(Visitor):
    """
    Implement each operation declared by Visitor. Each operation
    implements a fragment of the algorithm defined for the corresponding
    class of object in the structure. ConcreteVisitor provides the
    context for the algorithm and stores its local state. This state
    often accumulates results during the traversal of the structure.
    """

    def visit_concrete_element_a(self, concrete_element_a):
        pass

    def visit_concrete_element_b(self, concrete_element_b):
        pass


def main():
    concrete_visitor_1 = ConcreteVisitor1()
    concrete_element_a = ConcreteElementA()
    concrete_element_a.accept(concrete_visitor_1)


if __name__ == "__main__":
    main()
```

下面这个python例子说明，由访问者决定采用何种方法访问数据。

```python
class Node(object):
    pass


class A(Node):
    pass


class B(Node):
    pass


class C(A, B):
    pass


class Visitor(object):

    def visit(self, node, *args, **kwargs):
        meth = None
        for cls in node.__class__.__mro__:
            meth_name = 'visit_' + cls.__name__
            meth = getattr(self, meth_name, None)
            if meth:
                break

        if not meth:
            meth = self.generic_visit
        return meth(node, *args, **kwargs)

    def generic_visit(self, node, *args, **kwargs):
        print('generic_visit ' + node.__class__.__name__)

    def visit_B(self, node, *args, **kwargs):
        print('visit_B ' + node.__class__.__name__)


a = A()
b = B()
c = C()
visitor = Visitor()
visitor.visit(a)
visitor.visit(b)
visitor.visit(c)
```


### 责任链模式（Chain of Responsibility）☆☆☆☆

责任链模式（Chain of Responsibility Pattern）为请求创建了一个接收者对象的链。这种模式给予请求的类型，对请求的发送者和接收者进行解耦。这种类型的设计模式属于行为型模式。

注解： 责任链模式在python中也很常见，例如 Django中的中间件、scrapy中的中间件： 通过层层包裹来实现链式处理。形式有很大区别，思想可以认为是差不多的。感觉python的实现方式更优雅。


下面这个例子，纯粹的链式传递。 **很多时候，没必要照搬例子，理解设计模式的思想就可以了， 形而上**

```python
import abc


class Handler(metaclass=abc.ABCMeta):
    def __init__(self, successor=None):
        self._successor = successor

    @abc.abstractmethod
    def handle_request(self):
        pass


class ConcreteHandler1(Handler):
    def handle_request(self):
        if True:  # if can_handle:
            pass
        elif self._successor is not None:
            self._successor.handle_request()


class ConcreteHandler2(Handler):
    """
    Handle request, otherwise forward it to the successor.
    """

    def handle_request(self):
        if False:  # if can_handle:
            pass
        elif self._successor is not None:
            self._successor.handle_request()


def main():
    concrete_handler_1 = ConcreteHandler1()
    concrete_handler_2 = ConcreteHandler2(concrete_handler_1)
    concrete_handler_2.handle_request()


if __name__ == "__main__":
    main()
```

### 备忘录模式（Memento） ☆☆

备忘录模式（Memento Pattern）保存一个对象的某个状态，以便在适当的时候恢复对象。备忘录模式属于行为型模式。

注解： 这算是基本的模式了吧。 对象提供保存状态和恢复状态的能力。 例如 神经网络提供保存模型和恢复模型的能力。 这也叫模式？？？？

```python
import pickle


class Originator:
    """
    Create a memento containing a snapshot of its current internal
    state.
    Use the memento to restore its internal state.
    """

    def __init__(self):
        self._state = None

    def set_memento(self, memento):
        previous_state = pickle.loads(memento)
        vars(self).clear()
        vars(self).update(previous_state)

    def create_memento(self):
        return pickle.dumps(vars(self))


def main():
    originator = Originator()
    memento = originator.create_memento()
    originator._state = True
    originator.set_memento(memento)


if __name__ == "__main__":
    main()
```

### 解释器模式（Interpretor） ☆

解释器模式（Interpreter Pattern）提供了评估语言的语法或表达式的方式，它属于行为型模式。这种模式实现了一个表达式接口，该接口解释一个特定的上下文。这种模式被用在 SQL 解析、符号处理引擎等。


注解： 可以理解成为翻译型代理么。。。 SQL 解析 确实是一处主要的应用场景。

其他能用的场景非常少。 我想更多的是命令的转译，也算是一种 代理 模式？

```python
import abc


class AbstractExpression(metaclass=abc.ABCMeta):
    """
    Declare an abstract Interpret operation that is common to all nodes
    in the abstract syntax tree.
    """

    @abc.abstractmethod
    def interpret(self):
        pass


class NonterminalExpression(AbstractExpression):
    """
    Implement an Interpret operation for nonterminal symbols in the grammar.
    """

    def __init__(self, expression):
        self._expression = expression

    def interpret(self):
        self._expression.interpret()


class TerminalExpression(AbstractExpression):
    """
    Implement an Interpret operation associated with terminal symbols in
    the grammar.
    """

    def interpret(self):
        pass


def main():
    abstract_syntax_tree = NonterminalExpression(TerminalExpression())
    abstract_syntax_tree.interpret()


if __name__ == "__main__":
    main()
```
